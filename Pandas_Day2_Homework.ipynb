```python
import requests
```

```python
data = requests.get('https://api.github.com/repos/pandas-dev/pandas/issues')
```

```python
data = data.json()
```

```python
try:
    users = []
    for i in range(len(data)):
        users.append(data[i]['user'])
except Exception as e:
    print("Error in creating users list : ", e)
```

```python
df1 = pd.DataFrame(data,columns=['url','repository_url','labels_url','user'])
```

```python
df1.head()
```

```python
try:
    df2 = pd.DataFrame(users)
except Exception as e:
    print("Error in creating the users Dataframe : ", e)
```
